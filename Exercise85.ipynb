```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df=pd.read_csv("Dummy_Dataset_StudentHappinessScore.csv")
df.head()
```

```python
life_ladder = df['Life_Ladder']
perception_of_academics = df['Perception_of_Academics']
generosity = df['Generosity']
```

```python
columns = [life_ladder, perception_of_academics, generosity]
column_labels = ['Life Ladder', 'Perception of Academics', 'Generosity']

# Create separate 3D scatter plots for each column
for i, column in enumerate(columns):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Create the 3D scatter plot with continuous color mapping
    scatter = ax.scatter(life_ladder, perception_of_academics, generosity, c=column, cmap='viridis')

    # Add color bar to indicate the mapping of the current column's values
    cbar = fig.colorbar(scatter, ax=ax, label=column_labels[i])

    ax.set_xlabel('Life Ladder')
    ax.set_ylabel('Perception of Academics')
    ax.set_zlabel(column_labels[i])
    ax.set_title(f'Student Happiness Visualization - {column_labels[i]}')

    plt.show()
```

This visualization can be used to identify clusters of data points. The points in each cluster will have the same color. This can be helpful for understanding the data and for selecting appropriate parameters for clustering algorithms.
When you color-code data points in a scatter plot based on a variable, it can help you visually identify clusters or patterns in the data. Points with similar colors tend to have similar values for the chosen variable.
