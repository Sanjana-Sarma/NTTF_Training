```python
def remove_outliers(data, column):
    Q1 = np.percentile(data[column], 25)
    Q3 = np.percentile(data[column], 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# Remove outliers from 'hrs_studied' and 'grades' columns
df1 = pd.read_csv('Test_grade_outliers - Sheet1.csv')
df1_C = remove_outliers(df1, 'hrs_studied')
df1_C = remove_outliers(df1_C, 'grade')

# Extract cleaned 'hrs_studied' and 'grades' columns
x1_c = df1_C['hrs_studied'].values.reshape(-1,1)
y1_c = df1_C['grade'].values
degree = 3
poly_features = PolynomialFeatures(degree=degree, include_bias=False)
x1_c_poly = poly_features.fit_transform(x1_c)
import sklearn.linear_model as lm
lr1_c=lm.LinearRegression()
lr1_c.fit(x1_c_poly, y1_c)
print('Intercept: ', lr1_c.intercept_)
print('Coefficient: ', lr1_c.coef_)
predicted_grades_c = lr1_c.predict(x1_c_poly)
df1_C['predicted_grades']=predicted_grades_c
print(df1_C)
test_hr = int(input("Enter number of hours studied for which grade is to be predicted: "))
test_hr_poly=poly_features.transform(np.array([[test_hr]]))
predict_grade = lr1_c.predict(test_hr_poly)
print('Using predict function: ', predict_grade[0])
r2 = r2_score(y1_c, predicted_grades_c)
print("R-Squared Value: ", r2)
mae = mean_absolute_error(y1_c, predicted_grades_c)
print("Mean Absolute Error Value: ", mae)
rmse = np.sqrt(mean_squared_error(y1_c, predicted_grades_c))
print("RMSE Value: ", rmse)
```
