```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
print(cancer.target_names)
```

```python
print(cancer['feature_names'])
```

```python
X=cancer.data
y=cancer.target
```

```python
f=plt.figure(1)
plt.scatter(X[:,23],X[:,27],c=y)
plt.xlabel('Worst Area')
plt.ylabel('Worst Concave Points')
plt.title('Original data')
f.show()
```

```python
clf=PCA()
clf.fit(X,y)
explained_variance = clf.explained_variance_ratio_
print(explained_variance)
print(clf.singular_values_)

plt.plot(range(1, len(explained_variance) + 1), explained_variance.cumsum(), marker='o', linestyle='--')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance vs. Number of Components')
plt.show()
```

```python
pca=decomposition.PCA(n_components=4)#n_components depends on the elbow point from previous explained variance ratio graph
pca.fit(X)
X1=pca.transform(X)
```

```python
g=plt.figure(2)
plt.scatter(X1[:,0],X1[:,1],c=y)
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title('PCA data')
```
