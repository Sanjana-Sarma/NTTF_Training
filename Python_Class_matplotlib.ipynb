```python
import matplotlib.pyplot as plt
plt.axis([0,6,0,30])
plt.title('My first plot')
plt.xlabel('counting')
plt.ylabel('squared values')
plt.plot([1,2,3,4,5], [1,4,9,16,25], 'bo')
```

```python
import matplotlib.pyplot as plt
plt.axis([0,6,0,30])
plt.title('My first plot', fontsize = 22, fontname = 'Times New Roman')
plt.xlabel('Counting', color = 'gray')
plt.ylabel('Squared Values', color = 'gray')
plt.plot([1,2,3,4,5], [1,4,9,16,25], 'bo')
```

```python
import matplotlib.pyplot as plt
plt.axis([0,6,0,30])
plt.title('My first plot', fontsize = 22, fontname = 'Times New Roman')
plt.xlabel('Counting', color = 'gray')
plt.ylabel('Squared Values', color = 'gray')
plt.text(1,1.5, 'First')
plt.text(2,4.5, 'Second')
plt.text(3,9.5, 'Third')
plt.text(4, 16.5, 'Fourth')
plt.text(5, 25.5, 'Fifth')
plt.plot([1,2,3,4,5], [1,4,9,16,25], 'bo')
```

```python
import matplotlib.pyplot as plt
plt.axis([0,10,0,30])
plt.title('My first plot', fontsize = 22, fontname = 'Times New Roman')
plt.xlabel('Counting', color = 'gray')
plt.ylabel('Squared Values', color = 'gray')
plt.text(1,1.5, 'First')
plt.text(2,4.5, 'Second')
plt.text(3,9.5, 'Third')
plt.text(4, 16.5, 'Fourth')
plt.text(5, 25.5, 'Fifth')
plt.text(1.1,25,r'$y=x^2$', fontsize = 18, bbox = {'facecolor':'red', 'alpha':0.2})
plt.text(1.1,22,r'$y=x/2$', fontsize = 18, bbox = {'facecolor':'blue', 'alpha':0.2})
plt.text(1.1,19,r'$y=x*2$', fontsize = 18, bbox = {'facecolor':'green', 'alpha':0.2} )
plt.grid(True)
plt.plot([1,2,3,4,5], [1,4,9,16,25],'ro')
plt.plot([1,2,4,6,8],[0.5,1,2,3,4],'*','bo')
plt.plot([1,3,3.5,4,4.5],[2,6,7,8,9],'^','green')
plt.legend(['First Series', 'Second Series','Third Series'], loc = 1)
```
