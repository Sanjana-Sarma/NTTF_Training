```r
df<-read.table("https://stat4ds.rwth-aachen.de/data/UN.dat", header=TRUE)
```

```r
head(df, 5)
```

```r
matrix(cbind(mean(df$GDP),sd(df$GDP),
            mean(df$HDI), sd(df$HDI),
            mean(df$GII), sd(df$GII),
            mean(df$Fertility), sd(df$Fertility),
            mean(df$CO2), sd(df$CO2),
            mean(df$Homicide), sd(df$Homicide),
            mean(df$Prison), sd(df$Prison),
            mean(df$Internet), sd(df$Internet)), nrow=8)
```

```r
pairs(~GDP+HDI+GII+Fertility+CO2+Homicide+Prison+Internet, data=df)
```

```r
model_gdp_hdi <- lm(GDP~HDI, data=df)
model_gdp_gii <- lm(GDP~GII, data=df)
model_gdp_fertility <- lm(GDP~Fertility, data=df)
model_gdp_co2 <- lm(GDP~CO2, data=df)
model_gdp_homicide <- lm(GDP~Homicide, data=df)
model_gdp_prison <- lm(GDP~Prison, data=df)
model_gdp_internet <- lm(GDP~Internet, data=df)
```
